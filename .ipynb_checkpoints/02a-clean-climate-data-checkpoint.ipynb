```python
import pandas as pd
import shutil
```

```python
raw = pd.read_csv("DATA/RAW DATA/Climate Change Trends.csv")
print(raw)
```

```python
# Focus on core environmental and population indicators that could influence TB dynamics. We drop other columns.
clean = raw.drop(['Renewable Energy (%)', 'Forest Area (%)', 'Sea Level Rise (mm)'], axis=1)
print(clean)
```

```python
# Sort the DataFrame by the 'Year' column in ascending order
clean_sorted = clean.sort_values(by='Year')
print(clean_sorted)
```

```python
#countries appear multiple times for a given year. This is probably due to updates, corrections, or multiple variables per year. 
# In this case we will aggregate multiple entries into one by averaging the numeric-values per country/year.  
clean_climate_avg = clean.groupby(['Country', 'Year'], as_index=False).mean()
print(clean_climate_avg)
```

```python
# Save clean_climate_avg to csv file
clean_climate_avg.to_csv("clean_climate_data.csv", index=False)
```

```python
# Load the clean csv file to DATA/CLEAN DATA
source = "clean_climate_data.csv"
destination = "DATA/CLEAN DATA"
shutil.move(source, destination)
```
