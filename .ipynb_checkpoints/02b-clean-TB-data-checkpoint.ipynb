```python
import pandas as pd
import shutil
```

```python
raw = pd.read_csv("DATA/RAW DATA/Tuberculosis Trends.csv")
print(raw)
```

```python
# select columns to drop
print(raw.columns)
```

```python
# Filter the original DataFrame raw to keep only the columns relevant to tuberculosis and its risk factors, columns needed for this analysis.
clean = raw.loc[:, ['Country', 'Year', 'TB_Cases', 'TB_Deaths', 'TB_Incidence_Rate', 'TB_Mortality_Rate',
                  'Population',  'Urban_Population_Percentage', 'Malnutrition_Prevalence', 'Smoking_Prevalence', 
                  'TB_Doctors_Per_100K', 'Access_To_Health_Services', 'BCG_Vaccination_Coverage']]
print(clean)
```

```python
# # Sort the DataFrame by the 'Year' column in ascending order
clean = clean.sort_values(by='Year')
print(clean)
```

```python
# Average duplicates for each country and year. To have each country appear once a year.
clean_TB_avg = clean.groupby(['Country', 'Year'], as_index=False).mean()
print(clean_TB_avg)
```

```python
clean_TB_avg.head(50)
```

```python
#clean_TB_avg to csv
clean_TB_avg.to_csv('clean-TB-data.csv', index=False)
```

```python
# Load the clean-TB-data.csv to DATA/CLEAN DATA
source = "clean-TB-data.csv"
destination = "DATA/CLEAN DATA"
shutil.move(source, destination)
```
