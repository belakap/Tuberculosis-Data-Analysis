```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw = pd.read_csv("DATA/RAW DATA/Climate Change Trends.csv")
print(raw)
```

```python
# List of columns to plot
columns = [
    'Avg Temperature (°C)', 'Sea Level Rise (mm)', 'Rainfall (mm)', 'Population',
    'Extreme Weather Events', 'Forest Area (%)'
]

# Plot each histogram in a separate figure
for col in columns:
    plt.figure(figsize=(6, 4))
    raw[col].hist(bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {col}', fontsize=14)
    plt.xlabel(col, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.tight_layout()
    plt.show()
```

```python
# Group by country and compute the mean for each variable
country_summary = raw.groupby('Country')[[
    'Avg Temperature (°C)',
    'Sea Level Rise (mm)',
    'Rainfall (mm)',
    'Population',
    'Extreme Weather Events',
    'Forest Area (%)'
]].mean().reset_index()
print(country_summary)
```

```python
variables = [
    'Avg Temperature (°C)', 
    'Sea Level Rise (mm)', 
    'Rainfall (mm)', 
    'Population', 
    'Extreme Weather Events', 
    'Forest Area (%)'
]
country_stats = raw.groupby('Country')[variables].mean().reset_index()
hottest = country_stats.nlargest(10, 'Avg Temperature (°C)')
most_sea_rise = country_stats.nlargest(10, 'Sea Level Rise (mm)')
most_rainfall = country_stats.nlargest(10, 'Rainfall (mm)')
most_extreme_weather = country_stats.nlargest(10, 'Extreme Weather Events')
least_forest = country_stats.nsmallest(10, 'Forest Area (%)')
most_forest = country_stats.nlargest(10, 'Forest Area (%)')
```

```python
affected_countries = {
    'Hottest': hottest['Country'].tolist(),
    'Most Sea Level Rise': most_sea_rise['Country'].tolist(),
    'Most Rainfall': most_rainfall['Country'].tolist(),
    'Most Extreme Weather': most_extreme_weather['Country'].tolist(),
    'Least Forest Cover': least_forest['Country'].tolist(),
    'Most Forest Cover': most_forest['Country'].tolist()
}
summary_df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in affected_countries.items()]))
print(summary_df)
```

```python

# Top 10 countries most affected by each variable
hottest = country_stats.nlargest(10, 'Avg Temperature (°C)')[['Country', 'Avg Temperature (°C)']]
most_sea_rise = country_stats.nlargest(10, 'Sea Level Rise (mm)')[['Country', 'Sea Level Rise (mm)']]
most_rainfall = country_stats.nlargest(10, 'Rainfall (mm)')[['Country', 'Rainfall (mm)']]
most_extreme_weather = country_stats.nlargest(10, 'Extreme Weather Events')[['Country', 'Extreme Weather Events']]
least_forest = country_stats.nsmallest(10, 'Forest Area (%)')[['Country', 'Forest Area (%)']]

# Print the results
print("\n🌡️ Top 10 Hottest Countries:")
print(hottest)

print("\n🌊 Top 10 Countries with Most Sea Level Rise:")
print(most_sea_rise)

print("\n🌧️ Top 10 Countries with Most Rainfall:")
print(most_rainfall)

print("\n🌪️ Top 10 Countries with Most Extreme Weather Events:")
print(most_extreme_weather)

print("\n🌲 Bottom 10 Countries with Least Forest Cover:")
print(least_forest)
```
